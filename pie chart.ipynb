```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
val=[300,500,1600,450,590]
lab=['price','sell','total','shirt','sharee']
plt.pie(val,labels=lab)
```

```python
val=[300,500,1600,450,590]
lab=['price','sell','total','shirt','sharee']
plt.pie(val,labels=lab,radius=2)
```

```python
val=[300,500,1600,450,590]
lab=['price','sell','total','shirt','sharee']
plt.pie(val,labels=lab,radius=2,autopct='%.02f%%',shadow=True,startangle=45)
```

```python
val=[300,500,1600,450,590]
lab=['price','sell','total','shirt','sharee']
plt.pie(val,labels=lab,autopct='%0.3f%%',explode=[0.05,0,0.05,0,0])
```

# convert chart to image

```python
val=[300,500,1600,450,590]
lab=['price','sell','total','shirt','sharee']
plt.pie(val,labels=lab,radius=2,autopct='%.02f%%',shadow=True,startangle=45)
plt.savefig("piechat.jpg",Transparent=True,pad_inches=2)
```

```python
val=[300,500,1600,450,590]
lab=['price','sell','total','shirt','sharee']
plt.pie(val,labels=lab,radius=2,autopct='%.02f%%',shadow=True,startangle=45)
plt.savefig("piechat.pdf",Transparent=True)
```
