# Axes,label,legent and grid

```python
days=[1,2,3,4,5]
maxt=[50,60,70,80,90]
mint=[10,60,20,40,50]
avg=[30,40,50,60,70]
plt.plot(days,maxt,label='Max')
plt.plot(days,mint,label='Min')
plt.plot(days,avg,label='average')
plt.legend()
plt.grid()
#plt.legend(loc='upper right')
#plt.legend(loc='lower left')
plt.legend(loc='best')
plt.shadow=True
```
