```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
product=[1,2,3,4]
price=[15,25,40,30]
#profit=[5,7,10,5]
plt.bar(product,price)
```

```python
product=[1,2,3,4]
price=[15,25,40,30]
plt.bar(product,price,label='price',width=.5)
```

```python
product=[1,2,3,4]
price=[15,25,40,30]
plt.bar(product,price,label='price',color='red',width=.5)
```

```python
product=[1,2,3,4]
price=[15,25,40,30]
profit=[5,7,10,5]
plt.bar(product,price,label='price',color='red',width=.5)
plt.bar(product,profit,label='profit',width=.2)
plt.legend()
plt.shadow=True
#plt.grid(color='m')
```

# import showing different chart

```python
import numpy as np
```

```python
product=np.array([1,2,3,4])
price=[15,25,40,30]
profit=[5,7,10,5]
plt.bar(product-.2,price,label='price',width=0.5,color='red')
plt.bar(product+0.3,profit,label='profit',width=0.5)
plt.legend()
plt.savefig('barchart.png')
```
